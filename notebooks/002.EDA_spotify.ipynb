# **Charts**

## **Boxplot chart**

```python
df_without_unnamed = df.drop(columns=['Unnamed: 0'])
fig = plt.figure(figsize =(15, 7))
plt.boxplot(df_without_unnamed.select_dtypes(include=['float', 'int']).values, tick_labels=df_without_unnamed.select_dtypes(include=['float', 'int']).columns.values)
plt.show()
```

Podemos observar varias distribuciones interesantes.
* **Popularity:** se puede observar que la mayoría de pistas tienen popularidades un poco más bajo de 50 (si estamos hablando de un intervalo de 0 a 100, hablamos que la mayoría tienen entre la mitad y más bajo que la mitad su rango de popularidad).
* **Loudness:** observamos que la mayoría de pistas presentan su sonoridad por debajo de los 0 decibeles, por lo que podría significar que la mayoría de pistas pueden 'comportarse' como pistas más suaves y menos intensas. Se presentan valores atípicos pero son normales dado el contexto de decibeles (podemos tener canciones con decibeles muy bajos).
* **Tempo:** La mayoría de pistas tienen un tempo apróximado normal (un intervalo entre 100 y 150). Se tiene que investigar los valores atipicos como el tempo=0, ya que es algo raro que suceda esto, porque podría significar que no hay movimiento en la canción.
* **Duration_min:** La mayoría de pistas duran entre 0 y 5 minutos como se ve en la distribución. No se considera revisar los valores atipicos, pues una pista puede ser un podcast (de una hora o hora y media), sin embargo, se debe revisar duration_min=0

## **Map columns and explicit analysis chart**

Podemos observar que el 91,4% de las canciones no son explicitas.

```python
plt.bar(df['duration_grouped'].value_counts().index, df['duration_grouped'].value_counts().values, color='skyblue')
plt.title('Duration of the tracks')
plt.xlabel('Duration (in minutes)')
plt.ylabel('Number of tracks')
plt.show()
```

## **Heatmap**

```python
# Assuming 'df_without_unnamed' is my df without the 'Unnamed: 0' column
plt.figure(figsize=(30, 18))

# Using Seaborn to create a heatmap
sns.heatmap(df_without_unnamed.select_dtypes(include=['int', 'float']).corr(), fmt='.2f', annot=True)

plt.title('Correlation Heatmap')
plt.show()
```

* Correlación entre 'loudless' y 'energy': es claro que los decibeles influyen en si la música es más animada, calmada, etc.
* Correlación entre 'valence' y 'danceability': normalmente cuando se baila, se baila con canciones animadas. Sin embargo, sería algo interesante ver el comportamiento de bailabilidad en canciones negativas.
* Correlación entre 'acouticness' y 'energy': se puede ver interesante la correlación negativa que tiene, pues al tener canciones acusticas, ¿se puede caer en la categoría de una canción negativa? Se deben realizar más analisis de esta correlación.
* Correlación entre 'loudless' y 'acouticness': es claro que las canciones acusticas presentan decibeles más bajos, lo que puede indicar esta posible correlación no es nada fuera de lo común.
